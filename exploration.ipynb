```
import torch
import numpy as np
```

```
logits = torch.randn(4,10)
lbls = torch.ones(4)
print(logits.shape)
print(lbls.shape)
```

```
accuracy = 0
ps = torch.exp(logits) / (torch.sum(torch.exp(logits)))
top_p, top_class = ps.topk(1, dim=1)
```
